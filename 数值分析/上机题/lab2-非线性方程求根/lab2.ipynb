```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn

x = np.linspace(0, 50, 100)  # 生成 x 值的数组

# 计算第一类零阶贝塞尔函数
J_n = jn(0, x)

# 绘制第一类零阶贝塞尔函数曲线
plt.figure()
plt.plot(x, J_n, label=f'J0(x)')  # 第一类零阶贝塞尔函数
plt.xlabel('x')
plt.ylabel('Value')
plt.title(f'Bessel Function of Order 0')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# import math
# from math import sign
import numpy as np
from numpy import sign, absolute
def zeroin(ini_a, ini_b, f):
    a = ini_a
    b = ini_b
    fa = f(0, a)
    fb = f(0, b)
    if fa*fb > 0:
        print("error! must change sign!")
        return None
    
    c = a
    fc = fa
    d = b-c
    e = d

    while fb != 0:
        if fa*fb>0:
            a = c
            fa = fc
            d = b-c
            e = d
        if absolute(fa)<absolute(fb):
            c = b
            b = a
            a = c
            fc = fb
            fb = fa
            fa = fc

        m = 0.5*(a-b)
        tol = 2.0 * eps * max(absolute(b), 1.0)
        if absolute(m) <= tol or fb==0:
            return b
        if absolute(e) < tol or absolute(fc) <= absolute(fb):
            d = m
            e = m
        else:
            s = fb/fc
            if a == c:
                p = 2.0*m*s
                q = 1.0 - s
            else:
                q = fa/fc
                r = fb/fc
                p = s*(2.0*m*q*(q-r)-(b-a)*(r-1.0))
                q = (q-1.0)*(r-1.0)*(s-1.0)
            if p>0.0:
                q = -q
            else:
                p = -p
            if 2.0*p <3.0*m*q-absolute(tol*q) and p < absolute(0.5*e*q):
                e = d
                d = p/q
            else:
                d = m
                e = m
                
        c = b
        fc = fb
        if absolute(d) > tol:
            b += d
        else:
            b = b-sign(b-a)*tol
            
        fb = f(0, b)


# eps = np.float64(0.001)
h = np.float64(0.01)
low_bound = np.float64(0.0)
high_bound = np.float64(h)
minima = []
while(len(minima)<10):
    x = zeroin(low_bound, high_bound, jn)
    if x is None:
        high_bound += h
        continue
    minima.append(x)
    low_bound = x+h
    high_bound = low_bound + h
print(minima)

x = np.linspace(0, 50, 100)  # 生成 x 值的数组
J_n = jn(0, x)
plt.figure()
plt.plot(x, J_n, label=f'J0(x)')  # 第一类零阶贝塞尔函数
for minimum in minima:
    plt.text(minimum, 0, 'x', color='r', fontsize=12, ha='center', va='center')  # 在解的位置添加一个红色的 'x' 标记

plt.xlabel('x')
plt.ylabel('Value')
plt.title(f'Bessel Function of Order 0')
plt.legend()
plt.grid(True)
plt.show()
```
